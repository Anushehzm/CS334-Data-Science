```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DayLocator, DateFormatter
%matplotlib inline
import seaborn as sns
sns.set()
from calendar import month_name, day_name
```

```python
df = pd.read_pickle('weather_clean.pkl')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
axes = pd.plotting.scatter_matrix(df, figsize=(23,23))
for ax in axes.flatten():
    ax.set_xlabel(ax.get_xlabel(), rotation=90)
    ax.set_ylabel(ax.get_ylabel(), rotation=0);
```

We can observe a lot of things from this plot. 

- Cloud Cover shows 'banding' revealing that it is an integer variable (a few others also show this)
- Gust speed is not affected by any other variable except wind speed
- The min, mean and max of several variables are almost perfectly identical to each other (straight line of $y=x$)
- Temperature is negatively affected by sea level pressure
- Same for wind direction and humidity
- Dew point has interesting almost linear relationships with humidity and sea level pressure
- A few variables (temperature, humidity) are normally distributed. 
- Wind Direction has a bimodal distribution. 
- Visibility and Precipitation are very heavily skewed.

Let's explore a couple of these observations in more detail. 

```python
sns.scatterplot(data=df, x='Max.Gust.SpeedMPH', y='Max.Wind.SpeedMPH');
```

We can see that after removing the 2 outliers these two are pretty much linearly correlated. The following regression line shows this relationship and as expected the varaince only increases significantly near the outliers. With this we can be sure that these variables are pretty good predictors of each other. 

```python
sns.regplot(data=df, x='Max.Gust.SpeedMPH', y='Max.Wind.SpeedMPH');
```

```python
corr_mat = df.corr()

_, ax = plt.subplots(figsize=(10,8))
sns.heatmap(corr_mat, cmap='RdBu_r', center=0, ax=ax);
```

The correlation heatmap reinforces the idea that the min, mean and max features are correlated with each others. So adding these features to a linear prediciton model will not be a good idea (multi-colinearity). 

Notice that a diverging color pallete is being used here to highlight both extremes since a very positive or a very negative correlation coefficient are both important.

To identify other patterns let's clear the noise a bit. 

```python
subset = [c for c in df.columns if not ('Min' in c or 'Max' in c)]
corr_mat = df[subset].corr()

_, ax = plt.subplots(figsize=(10,8))
sns.heatmap(corr_mat, cmap='RdBu_r', center=0, ax=ax);
```

Now the picture becomes more obvious. `CloudCover`, `PrecipitationIn` and `Mean.Humidity` are correlated with each other positively and negatively with `Mean.VisibilityMiles` which makes complete sense. Dew point and temperature are highly positively correlated. Dew point and humidity also have some positive correlation.

After this step we can be fairly certain which variables will pe good predictors of the other ones if we need to create a predictive model. 

```python
sns.lmplot(data=df, x='Mean.Humidity', y='Mean.VisibilityMiles', 
            col='CloudCover', col_wrap=5, )#color='steelblue');
```

The heatmap shows that there was is a negative linear relationship between humidity and visibility and a positive relationship between cloud cover and humidity and a negative relationship between cloud cover and visibility. All of these patterns can be seen in this faceted scatter plot. The use of small multiples has revealed a pattern here which would have been buried in a global scatter plot. 

The negative relationship between humidity and visibility only kicks in for cloud cover of level 8. Before that there is no trend. The effect of cloud cover on both numeric variables can also be seen as it pulls the points to the bottom and to the right as cloud cover increases.

```python
_, ax = plt.subplots(figsize=(15, 8))
sns.lineplot(data=df, x='Date', y='Mean.TemperatureF', ax=ax);
ax.axhline(y=df['Mean.TemperatureF'].mean(), color='black', linestyle='dashed', label='Mean');
ax.axhline(y=df['Mean.TemperatureF'].median(), color='gray', linestyle='dashed', label='Median');
ax.axhspan(ymin=df['Mean.TemperatureF'].mean()-df['Mean.TemperatureF'].std(), 
           ymax=df['Mean.TemperatureF'].mean()+df['Mean.TemperatureF'].std(),
           color='gray', alpha=0.4, label='$\mu \pm \sigma$')

ax.xaxis.set_major_locator(DayLocator(1))
ax.xaxis.set_major_formatter(DateFormatter('%d %b %y'))
ax.legend();
```

```python
weekly_temps = df[['Date','Min.TemperatureF','Mean.TemperatureF','Max.TemperatureF']].set_index('Date').rolling(7).mean()
monthly = df[['Date','Min.TemperatureF','Mean.TemperatureF','Max.TemperatureF']].set_index('Date').rolling(30).mean()

_, ax = plt.subplots(figsize=(15, 8))
sns.lineplot(data=df, x='Date', y='Mean.TemperatureF', alpha=0.7, ax=ax);
sns.lineplot(x=weekly_temps.index, y=weekly_temps['Mean.TemperatureF'], label='Weekly Rolling Mean', ax=ax);
sns.lineplot(x=monthly.index, y=monthly['Mean.TemperatureF'], label='Monthly Rolling Mean', ax=ax);

ax.xaxis.set_major_locator(DayLocator(1))
ax.xaxis.set_major_formatter(DateFormatter('%d %b %y'))
ax.legend();
```

Here the orange line is the weekly rolling mean of the data (which is still in blue). A window of 7 days is kept and the mean of this window is plotted on the 7th day (which is why the first 6 values are missing from this plot). This trend is smoother than the original data since a lot of the erraticness in daily readings is gone. We can still see an up-down pattern (could be within a month?). The green line is the mothly rolling average where we have lost pretty much all the noise and are left with just the seasonal component of summer and winter. 

```python
monthly_temps = df.groupby(df['Date'].dt.month).mean()['Mean.TemperatureF']

_, ax = plt.subplots(figsize=(10,7))
ax.scatter(monthly_temps.index, monthly_temps.values, marker='D')
plt.xticks(range(1,13), [month_name[n] for n in range(1,13)], rotation=90);
```

The change of seasons is pretty clear from this picture. 

```python
dow_temps = df.groupby(df['Date'].dt.dayofweek).mean()['Mean.TemperatureF']

_, ax = plt.subplots(figsize=(10,7))
ax.scatter(dow_temps.index, dow_temps.values, marker='+')
plt.xticks(ticks=range(7), labels=[day_name[n] for n in range(7)], rotation=90);
```

There is somehow a trend within weeks as well, Wednesday being the hottest day and Saturday being the coldest.

```python
subset = df.set_index('Date')['2015-06-01':'2015-08-01'].reset_index()

_, ax1 = plt.subplots(figsize=(20, 8))
sns.lineplot(data=subset, x='Date', y='Mean.Humidity', label='Mean Humidity', color='C0', ax=ax1);

ax2 = ax1.twinx()
sns.lineplot(data=subset, x='Date', y='Mean.VisibilityMiles', label='Mean Visibility', color='C1', ax=ax2);

ax1.xaxis.set_major_locator(DayLocator(interval=5))
ax1.xaxis.set_major_formatter(DateFormatter('%d %b %y'))
ax1.legend(loc='upper left');
ax2.legend(loc='upper right');
```

Since the scales of humidity and visibility differ so much (x10 difference), they have been overlayed with independant y axes for clarity.

To make the trend clearer I have zoomed into the June - August period. It can be seen that there is a negative correlation between these two trends. The peaks of one coincide with the valleys of the other. 

```python
def pick_max_min(df,col,asc):
    return df.sort_values(by=col,ascending=asc).iloc[0]
monthly_max_temps = df.groupby(df['Date'].dt.month).apply(pick_max_min,"Mean.TemperatureF",False)['Mean.TemperatureF']
monthly_min_temps = df.groupby(df['Date'].dt.month).apply(pick_max_min,"Mean.TemperatureF",True)['Mean.TemperatureF']
```

```python
_, ax = plt.subplots(figsize=(10,7))
sns.lineplot(monthly_max_temps.index, monthly_max_temps.values, marker='D', color='C0', label='Max Mean Temperature')
sns.lineplot(monthly_min_temps.index, monthly_min_temps.values, marker='D', color='C1', label='Min Mean Temperature')
sns.lineplot(monthly_temps.index, monthly_temps.values, marker='D', color='C2', label='Average Mean Temperature')
ax.legend();
plt.xlabel("Months")
plt.ylabel("Temperature (F)")
plt.xticks(range(1,13),[month_name[n] for n in range(1,13)], rotation=90);
```

Here I have plotted temperature of hotest and coldest days along with the average monthly temperature we found out earlier. We can see that same trend holds in all three graphs. We see coldest day in February and hottest day in September.

```python
df['Month'] = df['Date'].dt.month
Events_pivot = df.pivot_table(
    index='Month', 
    columns='Events', 
    values='Mean.TemperatureF', 
    aggfunc= lambda x: len(x), # count
)
Events_pivot
```

When we pivot our dataframe around Events column, we are seeing NaN values for different entries. NaNs appear when month had no day corresponding to that event. We saw in part 1 of assignment how we can handle NaN values. Let's apply that here.

```python
Events_pivot[Events_pivot.isna()] = 0
Events_pivot
```

we can also draw a stacked bar chart insead of a line plot for the following.

```python
_, ax = plt.subplots(figsize=(10,7))
cols = Events_pivot.columns
for i in range(len(cols)):
    sns.lineplot(Events_pivot[cols[i]].index, Events_pivot[cols[i]].values, marker='D', label=cols[i])
ax.legend();
plt.xlabel("Months")
plt.ylabel("Occurrences")
plt.xticks(range(1,13),[month_name[n] for n in range(1,13)], rotation=90);
```

There is too much information to absorb here, which is never a good idea when visualising data. We can see that for a lot of categories graph stays close to zero. Lets ignore these categories and focus on events that can be displayed in more meaningful way.

```python
_, ax = plt.subplots(figsize=(10,7))
cols = ['Snow','Rain','None','Fog']
for i in range(len(cols)):
    sns.lineplot(Events_pivot[cols[i]].index, Events_pivot[cols[i]].values, marker='D', label=cols[i])
ax.legend();
plt.xlabel("Months")
plt.ylabel("Occurrences")
plt.xticks(range(1,13),[month_name[n] for n in range(1,13)], rotation=90);
```
